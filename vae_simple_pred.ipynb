## Adoption VAE to predict heterogeneity of reactor core

```python
from torch import nn
import torch.nn.functional as F
from torch.utils.tensorboard import SummaryWriter
```

```python
device = (
    "cuda"
    if torch.cuda.is_available()
    else "cpu"
)
```

```python
por = next(iter(tr))[0]
por = por.to(device)
por
```

```python
class VAE(nn.Module):
    def __init__(
        self,
        input_features
    ):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_features, 128),
            nn.ReLU()
        )
        self.mu, self.logvar = nn.Linear(128, 40), nn.Linear(128, 40)
        
        self.decoder = nn.Sequential(
            nn.Linear(40, 128),
            nn.ReLU(),
            nn.Linear(128, 20),
            # nn.Tanh()
        )

    def reparameterize(self, mu, logvar):
        std = torch.exp(logvar/2)
        z = mu + std*torch.rand_like(std)
        return z
    
    def forward(self, x):
        encoded = self.encoder(x)
        
        mu = F.relu(self.mu(encoded))
        logvar = F.relu(self.logvar(encoded))

        z = self.reparameterize(mu, logvar)

        decoded = self.decoder(z)
        return decoded, z, mu, logvar
        
```

```python
def loss_function(
    x,
    output,
    logvar,
    mu,
    batch_size
):
    recon_loss = F.mse_loss(output, x, reduction='sum') / batch_size
    kl_loss = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())

    return recon_loss + 0.002  * kl_loss
```

```python
def train(
    model,
    dataset,
    optimizer,
    loss_function
):
        epoch_loss  = 0
        for nbatch, (X,y) in enumerate(dataset):
            X,y = X.to(device), y.to(device)
            
            # Zero the gradients
            optimizer.zero_grad()
            
            
            output, z, mu, logvar = model(X)

            loss = loss_function(
                y,
                output,
                logvar,
                mu,
                X.shape[0]
            )
    
            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()
            
        print(f"loss: {epoch_loss}")

def test(
    model,
    dataset,
    epoch_num,
    writer
):
    test_loss = 0.0
    
    with torch.no_grad():
        for nbatch, (X,y) in enumerate(dataset):
            X,y = X.to(device), y.to(device)
            output, z, mu, logvar = model(X)
            
            test_loss += loss_function(
                y,
                output,
                logvar,
                mu,
                X.shape[0]
            ).item()

        # if epoch_num % 2 == 0:
        
        writer.add_scalar('Test/Loss', test_loss, global_step=epoch_num)
        writer.add_scalar('Test/Predicted', output[0].mean(), global_step=epoch_num)
        writer.add_scalar('Test/Original', y[0].mean(), global_step=epoch_num)

        writer.add_scalar('Test/Pr_O', y[0].mean(), global_step=output[0].mean())

            
        test_loss = test_loss / len(dataset)
        
        print(f"Test Error: \n Accuracy: Avg loss: {test_loss:>8f} \n")
    return z
```

```python
model = VAE(features).to(device)
```

```python
writer = SummaryWriter(f'irt_het/test1')
```

```python
adam = torch.optim.Adam(model.parameters())
```

```python
epoch = 100
for e in range(epoch):
    train(
        model,
        tr,
        adam,
        loss_function=loss_function
    )
    
    z = test(
        model,
        te,
        e,
        writer
    )
```
